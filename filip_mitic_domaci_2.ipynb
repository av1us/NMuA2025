```python
import numpy as np
import matplotlib.pyplot as plt
```

# Prvi zadatak

```python
l, I = np.loadtxt('Spektar Sunca.txt', delimiter=',', skiprows=2, unpack=True)
plt.plot(l, I, 'r', linewidth=2)
plt.xlabel('Talasna duzina [nm]')
plt.ylabel('Intenzitet zračenja')
plt.title('Spektar Sunčevog zračenja')
plt.show()

lmax = l[np.argmax(I)]
lmin = l[np.argmin(I)]
print('Maksimum intenziteta se nalazi na', lmax, 'nm, a minimum na', lmin, 'nm.')
```

# Drugi zadatak

```python
#pravimo matricu M-ova i e-ova
e=[0.05, 0.2, 0.5]
M=M = np.linspace(0, 2*np.pi, 10)
E = np.zeros_like(M)
e, M = np.meshgrid(e, M)
tacnost=1e-8

def kep(M, e, tacnost=1e-8):
    br=0
    E = np.copy(M)
    delta=2*tacnost
    while np.any(np.abs(delta) > tacnost):
        br+=1
        f=E-e*np.sin(E)-M
        fprim=1-e*np.cos(E)
        delta=f/fprim
        E=E-delta
    return E, br

print('rezultati:',kep(M,e)[0])
print('broj iteracija:', kep(M,e)[1])
```
