```python
import numpy as np
import matplotlib.pyplot as plt
```

## Prvi zadatak:

```python
def fib(n): #definisanje funkcije
    a = [0, 1]
    if n == 0:
        return []
    elif n == 1 or n == 2:
        return a[:n]
    else:
        for i in range(n-2):
            a.append(a[len(a)-2]+a[len(a)-1]) #svaki sledeći član posle prva dva je zbir prethodna dva
        return a

print(fib(10))
```

## Drugi zadatak:

```python
M = 3.6e31 #kg
G = 6.67430e-11 #m^3 kg^-1 s^-2
c = 3e8 #m/s
```

#### Definisanje funkcije i njenog izvoda:

```python
def fun(R):
    return (2*M*G)/R - c**2
def funprim(R):
    return -(2*M*G)/R**2 
```

#### Rešavanje Njutnovom metodom:

```python
#pravljenje niza za x osu i funkcije za grafik
Rniz = np.linspace(5e3, 7e4, 1000)
fniz = (2*M*G)/Rniz - c**2

#pravljenje grafika
plt.plot(Rniz, fniz, color='indianred', linewidth=1)
plt.title(r'Funkcija $f(R)=\frac{2MG}{R} - c^{2}$.')
plt.grid()
plt.xlabel(r'$R\,[m]$')
plt.ylabel(r'f(R)')
full_fig = plt.gcf()

#definisanje njutnovog metoda
R = 1e4 #inicijalna vrednost
def njutn(R, tol=1e-6):
    delta = 1
    broj=0
    while np.abs(delta)>tol:
        f = fun(R)
        fprim= funprim(R)
        delta = f/fprim
        plt.plot(R, fun(R), 'o', color='black', markersize=3) #plotovanje svakog koraka na grafik
        R = R - delta
        broj+=1
    return R, broj
Rf, broj = njutn(R)

print('Dobijena vrednost radijusa: {}'.format(Rf), 'm')
print('Broj iteracija: {}'.format(broj))
plt.plot(Rf, fun(Rf), 'o', color='green', markersize=7) #plotovanje iteracija
```
